## Matplotlib example workflow

```python
# 0. import matplotlib as get it ready for plotting in Jupyter
import matplotlib.pyplot as plt
%matplotlib inline

#1. Prepare the data
x = [1,2,3,4]
y = [11,22,33,44]

#2. Setup plot
fig, ax = plt.subplots(figsize = (10,10))

#3. Plot data
ax.plot(x,y)

#4. Customize plot
ax.set(title = 'Simple Plot',
      xlabel = 'x-axis',
      ylabel = 'y-axis')

#5. Save & show (saving the whole figure)
fig.savefig('sample-plot.png')
```

```python
# Create some data
x = np.linspace(0,10,100) #returns 100 numbers from 0 to 10
```

```python
# Make a plot from a dictionary
nut_butter_prices = {'Almond butter': 10,
                     'Peanut butter': 8,
                      'Cashew butter':12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), height = nut_butter_prices.values()); #x,y
ax.set(title = "Erica's Nut Butter Store",
       ylabel = 'Price ($)');
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()));
```

```python
#Make some data
#randn - standart normal distribution ( Gaussian)
x = np.random.randn(1000)
fig,ax = plt.subplots()
ax.hist(x);
```

# 2 options for subplots

```python
#Subplot option 1
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows = 2,
                                         ncols = 2,
                                         figsize = (10,5))

#Plot to each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
#Subplots option 2
fig, ax = plt.subplots(nrows = 2,
                      ncols = 2,
                       figsize = (10,5))

#Plot to each subplot index

#row  = 0, col = 0
ax[0,0].plot(x,x/2); 
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(10000));
```

### Which one to use? (pyplot vs matplotlib OO method)

* When plotting something quickly, okay to use pyplot method
* When plotting something more advanced, use the OO method

```python
over_50 = heart_disease[heart_disease.age > 50]
over_50
```

```python
# Pyplot Method
over_50.plot(kind = 'scatter',
            x = 'age',
            y = 'chol',
            c = 'target');
```

```python
# OO Method mixed with pyplot

fig, ax = plt.subplots(figsize = (10,6))
over_50.plot(kind = 'scatter',
            x = 'age',
            y = 'chol',
            c = 'target',
            ax = ax);

ax.set_xlim([45,80]);
```

```python
# OO Method from scratch - aka Object Oriented
fig, ax = plt.subplots(figsize = (10,6))

#Plot the data
scatter = ax.scatter(x = over_50.age,
                    y = over_50.chol,
                    c = over_50.target)

#Customize
ax.set(title = 'Heart Disease and Cholesterol Levels',
      xlabel = 'Age',
      ylabel = 'Cholesterol')

#Add a legend
ax.legend(*scatter.legend_elements(), title = 'Target');

#Add a horizontal line
ax.axhline(over_50.chol.mean(),
          linestyle = '--');

```

```python
over_50.head()
```

```python
#Subplots of chol, age, thalach

fig, (ax0,ax1) = plt.subplots(nrows = 2,
                            ncols = 1,
                            figsize = (10,10),
                             sharex = True)

#Add data to ax0

scatter = ax0.scatter(x = over_50.age,
                     y = over_50.chol,
                     c = over_50.target)

#Customize ax0
ax0.set(title = 'Heart Disease and Cholesterol Levels',
        xlabel = 'Age',
       ylabel = 'Cholesterol')

ax0.legend(*scatter.legend_elements(), title = 'Target')

ax0.axhline(y = over_50.chol.mean(),
           linestyle = '--');

# My part

mine_scatter = ax1.scatter(x = over_50.age,
                          y = over_50.thalach,
                          c = over_50.target)

ax1.set(title = 'Heart Disease and Thalach Levels',
       xlabel = 'Age',
       ylabel = 'Thalac')

ax1.legend(*scatter.legend_elements(),title = 'Target')

ax1.axhline(y = over_50.thalach.mean(), 
           linestyle = '--');

#Add a title to the figure

fig.suptitle('Heart Disease Analysis', fontsize = 16, fontweight = 'bold');
```

```python
#Set style
plt.style.use('seaborn-whitegrid')

# OO Method from scratch - aka Object Oriented
fig, ax = plt.subplots(figsize = (10,6))

#Plot the data
#Tem vrios esquemas de cores, s olhar
scatter = ax.scatter(x = over_50.age,
                    y = over_50.chol,
                    c = over_50.target,
                    cmap = 'winter') #this changes the color scheme

#Customize
ax.set(title = 'Heart Disease and Cholesterol Levels',
      xlabel = 'Age',
      ylabel = 'Cholesterol')

#Add a legend
ax.legend(*scatter.legend_elements(), title = 'Target');

#Add a horizontal line
ax.axhline(over_50.chol.mean(),
          linestyle = '--');
```

```python
plt.style.use('seaborn-whitegrid')

#Subplots of chol, age, thalach

fig, (ax0,ax1) = plt.subplots(nrows = 2,
                            ncols = 1,
                            figsize = (10,10),
                             sharex = True)

#Add data to ax0

scatter = ax0.scatter(x = over_50.age,
                     y = over_50.chol,
                     c = over_50.target,
                     cmap = 'winter')

#Customize ax0
ax0.set(title = 'Heart Disease and Cholesterol Levels',
        xlabel = 'Age',
       ylabel = 'Cholesterol')

ax0.set_xlim([50,80]) #tira a linha da frente
ax0.set_ylim([100,450])
ax0.legend(*scatter.legend_elements(), title = 'Target')

ax0.axhline(y = over_50.chol.mean(),
           linestyle = '--',
           color = 'r');

# My part

mine_scatter = ax1.scatter(x = over_50.age,
                          y = over_50.thalach,
                          c = over_50.target,
                          cmap = 'winter')

ax1.set(title = 'Heart Disease and Thalach Levels',
       xlabel = 'Age',
       ylabel = 'Thalac')

ax1.legend(*scatter.legend_elements(),title = 'Target')

ax1.axhline(y = over_50.thalach.mean(), 
           linestyle = '--',
           color = 'r');
ax1.set_ylim([60,200])

#Add a title to the figure

fig.suptitle('Heart Disease Analysis', fontsize = 16, fontweight = 'bold');
```
