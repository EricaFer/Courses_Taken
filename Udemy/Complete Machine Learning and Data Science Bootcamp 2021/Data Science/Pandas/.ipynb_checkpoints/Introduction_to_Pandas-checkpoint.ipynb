```python
#Column form series
seats_column = pd.Series([5, 5, 5, 5, 5])

#New Column called seats
car_sales['Seats'] = seats_column
car_sales
```

```python
car_sales['Seats'].fillna(5,inplace = True)
car_sales
```

```python
#Column from list

fuel_economy = [7.5,9.2,5.0,9.6,8.7,4.7,7.6,8.7,3.0,4.5]
car_sales['Fuel per 100KM'] = fuel_economy
car_sales

#OBS

#A serie n precisa ser do mesmo tamanho, ele enche com null
#a lista precisa ser do mesmo tamanho
```

```python
car_sales['Total Fuel Use (L)'] = car_sales['Odometer (KM)']/100 * car_sales['Fuel per 100KM']
car_sales
```

```python
#Create a Column from a single value
car_sales['Number of wheels'] = 4
car_sales
```

```python
car_sales['Passed road safety'] = True
car_sales
```

```python
car_sales.drop('Total Fuel Use (L)',axis = 1)
car_sales
```

```python
car_sales_shuffled = car_sales.sample(frac = 1) #fraction that will be randomized 1 = 100%
car_sales_shuffled
```

```python
#Only select 20% of the data 
car_sales_shuffled.sample(frac = 0.2)
```

```python
car_sales_shuffled.reset_index(drop = True, inplace = True)
car_sales_shuffled
```
